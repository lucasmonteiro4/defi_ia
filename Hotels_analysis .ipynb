```python
import pandas as pd
```

```python
hotels = pd.read_csv('features_hotels.csv')
hotels.head()
```

```python
hotels.isnull().sum()
```

```python
hotels.groupby('city')['hotel_id'].agg(['count']).sort_values(['count'], ascending=False)
```

```python
hotels.groupby('city')['hotel_id'].agg(['count']).sort_values(['count'], ascending=False).reset_index().plot(x='city', y='count', kind = 'bar', figsize = (10, 10))
```

```python
pool = hotels.groupby(
    ['city', 'pool']
)['hotel_id'].agg(['count'])

pool.unstack().sort_values(('count',0), ascending=False)
```

```python
pool_per = pool / hotels.groupby('city')['hotel_id'].agg(['count'])*100
pool_per.round(1).unstack().sort_values(('count',0), ascending=False)
```

```python
hotels.groupby(
    ['city', 'pool']
)['hotel_id'].count().unstack().plot(kind ='bar', figsize = (10, 10))
```

```python
parking = hotels.groupby(
    ['city', 'parking']
)['hotel_id'].agg(['count'])

pool.unstack().sort_values(('count',0), ascending=False)
```

```python
parking_per = parking / hotels.groupby('city')['hotel_id'].agg(['count'])*100
parking_per.round(1).unstack().sort_values(('count',0), ascending=False)
```

```python
children_policy = hotels.groupby(
    ['city', 'children_policy']
)['hotel_id'].agg(['count'])

children_policy.unstack().sort_values(('count',0), ascending=False)
```

```python
children_policy_per = children_policy / hotels.groupby('city')['hotel_id'].agg(['count'])*100
children_policy_per.round(1).unstack().sort_values(('count',0), ascending=False)
```

```python
group = hotels.groupby(
    ['city', 'group']
)['hotel_id'].agg(['count'])
group.unstack()
```

```python
group_per = group / hotels.groupby('city')['hotel_id'].agg(['count'])*100
group_per.round(1).unstack()
```
